#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about them.

```python
import pandas as pd

LEVEL_FIVE = "level5"

data_frame = pd.read_csv('assets/real_estate.csv', sep=';')

valdemorillo_data = data_frame.loc[data_frame[LEVEL_FIVE] == "Valdemorillo"]
galapagar_data = data_frame.loc[data_frame[LEVEL_FIVE] == "Galapagar"]

valdemorillo_price_mean = valdemorillo_data["price"].dropna().mean()
galapagar_price_mean = galapagar_data["price"].dropna().mean()

print(f"Average price in Valdemorillo: {valdemorillo_price_mean:.2f} USD")
print(f"Average price in Galapagar: {galapagar_price_mean:.2f} USD")

if valdemorillo_price_mean == galapagar_price_mean:
    conclusion = "The average prices of Valdemorillo and Galapagar are the same."
else:
    conclusion = "The average prices of Valdemorillo and Galapagar are different."

print(conclusion)
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

LEVEL_FIVE = "level5"
N_BINS = 10

data_frame = pd.read_csv('assets/real_estate.csv', sep=';')
south_belt_df = data_frame.loc[data_frame[LEVEL_FIVE].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])].copy()

south_belt_df['normalized_price'] = south_belt_df.groupby(LEVEL_FIVE)['price'].transform(lambda x: x / x.max())

fig, axes = plt.subplots(1, 4, figsize=(20, 6), sharey=True)

colors = ["purple", "blue", "green", "orange"]
populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

for i, population in enumerate(populations):
    subset = south_belt_df[south_belt_df[LEVEL_FIVE] == population]
    axes[i].hist(subset['normalized_price'].dropna(), bins=N_BINS, color=colors[i], alpha=0.7)
    axes[i].set_title(population)
    axes[i].set_xlabel("Normalized Price")

axes[0].set_ylabel("Frequency")
fig.suptitle("Normalized Price Distribution by Population")
plt.show()

```

Analysis of Normalized Price Histograms

The histograms of the populations Fuenlabrada, Leganés, Getafe, and Alcorcón reveal similar patterns in the distribution of normalized prices, with some notable differences:

Price Distribution in Each Population:

• Fuenlabrada: The histogram shows its highest peak between the ranges of 0.2 and 0.4, indicating that most properties fall within this relative price range. The remaining values are spread more uniformly, with less concentration in higher prices.
• Leganés: Here, the highest peak is around the 0.2 range, suggesting that a large number of properties are clustered at the lower end of the normalized price range. The other values are distributed evenly across the remaining ranges, with no additional concentrations in higher prices.
• Getafe: This histogram has two notable peaks, one in the 0.0 range and another at 0.4, with the latter being the highest. This suggests that Getafe has a wider price distribution, with some properties in the low range and others closer to the maximum relative price.
• Alcorcón: In this population, the highest peak is just below 0.2, indicating a tendency toward relatively lower prices. The remaining prices are spread out evenly across the rest of the range, with no significant concentrations in higher values.


• Comparison Between Populations:
    ◇ Overall, the histograms show that all four populations have prices mostly concentrated within the lower ranges of the normalized scale, meaning there is a predominance of lower-priced properties in each area.
    ◇ Getafe is the only population showing a slight tendency toward higher prices, as reflected by its second peak at 0.4.
    ◇ This suggests that while the four populations share a similar price range, Getafe may have a slightly higher proportion of properties at higher relative prices.


The histograms confirm that prices in the four populations are within a similar range, with a slight difference in Getafe, which has somewhat higher prices compared to the others.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
import pandas as pd
import matplotlib.pyplot as plt

LEVEL_FIVE = "level5"
N_BINS = 10

data_frame = pd.read_csv('assets/real_estate.csv', sep=';')
south_belt_df = data_frame.loc[data_frame[LEVEL_FIVE].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])].copy()
south_belt_df = south_belt_df[south_belt_df['surface'] > 0]
south_belt_df["pps"] = south_belt_df['price'] / south_belt_df['surface']

fig, axes = plt.subplots(1, 2, figsize = (16, 4), sharey = True)
 
colors = ["purple", "green"]
populations = ["Getafe", "Alcorcón"]
for i, population in enumerate(populations):
    subset = south_belt_df[south_belt_df[LEVEL_FIVE] == population]
    axes[i].hist(subset['pps'].dropna(), bins=N_BINS, color = colors[i], alpha = 0.7)
    axes[i].set_title(population)
    axes[i].set_xlabel("Price per square meter (EUR/m²)")

axes[0].set_ylabel("Frequency")
fig.suptitle("Price per square meter (EUR/m²) between Getafe and Alcorcón")
plt.show()
```

• Getafe: Tends to have a higher concentration of properties at a relatively affordable price per square meter, centered between 1500 and 2000 EUR/m².
• Alcorcón: Shows more dispersed and slightly higher prices in some cases, though at lower frequencies, particularly in the 2000 to 2500 EUR/m² range.

In Getafe, the properties tend to be more affordable on average than in Alcorcón, where price per square meter is more spread out and peaks at somewhat higher values. This suggests that in Getafe, there are more options at moderate prices, while in Alcorcón, prices can reach higher ranges, though less frequently.
